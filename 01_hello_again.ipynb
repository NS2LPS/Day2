# Exercises
The output of the QM is now connected to a spectrum analyzer. You can see in real time the pulses that you produce. In a typical experiment, a pulse is sent on the qubit, the qubit is measured and finally a long waiting time is inserted before the next run in order to let time to the qubit to relax.

You will use the following typical code sequence: 
```
nmax = 10
with program() as prog:
    n = declare(int)  # QUA variable for the loop 
    with infinite_loop_():
        # Infinite loop to repeat the experiment
        with for_(n, 0, n < nmax, n + 1):  # Experiment loop
            # Trigger the spectrum analyzer
            play('trigger','trigger')
            # Send a pulse
            play('pulse','qubit',duration=4*n)
            # Wait a long time before next iteration
            wait(100*u.us)
```
